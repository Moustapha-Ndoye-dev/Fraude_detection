# 03 - EDA segmentation client

Objectif: comprendre les profils clients, revenus, depenses, canaux d'achat et reponses marketing.

```python
from pathlib import Path
import sys
import pandas as pd

ROOT = Path('..').resolve()
sys.path.insert(0, str(ROOT / 'src'))
from ml_project.data.loaders import read_customer_data
from ml_project.features.customers import add_customer_features

raw = read_customer_data()
df = add_customer_features(raw)
df.head()
```

```python
# Valeurs manquantes et statistiques principales
display(df.isna().sum().sort_values(ascending=False).head(10))
display(df[['Income', 'Total_Spend', 'Total_Purchases', 'Response', 'Recency']].describe())
```

```python
# Revenus et depenses par education
display(df.groupby('Education')[['Income', 'Total_Spend', 'Response']].mean().sort_values('Total_Spend', ascending=False))
df['Total_Spend'].plot(kind='hist', bins=40, title='Distribution des depenses clients')
```

Conclusion metier: les variables de valeur client, canaux d'achat et reponse campagne doivent guider les segments finaux, pas seulement les variables demographiques.
